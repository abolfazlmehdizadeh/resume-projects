```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
# yellow color
color = int(cv.cvtColor(np.array([[[0, 255, 255]]], 'uint8'), cv.COLOR_BGR2HSV)[0][0][0])
```

```python
capture = cv.VideoCapture(1)

last_rame = None
while 1:
    ret, frame = capture.read()
    if not ret:
        break
    hsv_frame = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    
    lower = np.array([color - 10, 50, 50])
    upper = np.array([color + 10, 255, 255])
    

    mask = cv.inRange(hsv_frame, lower, upper)
    res = cv.bitwise_and(frame, frame, mask=mask)
    
    cv.imshow('res', res)
    
    if cv.waitKey(5) == 27:
        last_rame = res
        break

cv.destroyAllWindows()
capture.release()
plt.imshow(cv.cvtColor(last_rame, cv.COLOR_BGR2RGB))
```
