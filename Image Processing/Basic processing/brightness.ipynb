```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image = cv.imread('pic.jpg')
```

## Linear method

```python
plt.figure(figsize=[16, 16])
plt.subplot(141)
plt.title('a/2 - 5b')
plt.imshow(cv.convertScaleAbs(image, alpha=0.5, beta=-5))
plt.subplot(142)
plt.title('original')
plt.imshow(image)
plt.subplot(143)
plt.title('2a + 5b')
plt.imshow(cv.convertScaleAbs(image, alpha=2, beta=+5))
plt.subplot(144)
plt.title('4a + 5b')
plt.imshow(cv.convertScaleAbs(image, alpha=4, beta=+5))
```

## Gamma correction method

```python
def brightness_with_gamma(img, g):
    LT = np.empty((1, 256), np.uint8)
    for i in range(256):
        LT[0, i] = pow(i / 255, g) * 255
    return cv.LUT(img, LT)


plt.figure(figsize=[16, 16])
plt.subplot(141)
plt.title('gamma = 1.5')
plt.imshow(brightness_with_gamma(image, 1.5))
plt.subplot(142)
plt.title('gamma = 1 (original)')
plt.imshow(brightness_with_gamma(image, 1))
plt.subplot(143)
plt.title('gamma = 0.7')
plt.imshow(brightness_with_gamma(image, 0.7))
plt.subplot(144)
plt.title('gamma = 0.4')
plt.imshow(brightness_with_gamma(image, 0.4))
```

## Comparison
