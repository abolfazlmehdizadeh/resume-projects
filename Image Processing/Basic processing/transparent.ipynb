### import packages

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### read image and show

```python
image = cv.imread('1.png')
plt.imshow(image[...,::-1])
```

### get height and width and Covert image as bgr to rgba

```python
H, W = image.shape[0:2]
n_image = cv.cvtColor(image, cv.COLOR_BGR2RGBA)
n_image[...,3] = np.ones((H, W), np.uint8) * 20
```

### show image in RGBA with alpha 12%

```python
plt.imshow(n_image)
```

### crop and show image with alpha 100%

```python
crop_image = image[H//7 : H//3, W//2 : 3*W//5, ::-1]

    
n_image[H//7 : H//3, W//2 : 3*W//5, 3 ] = np.ones(crop_image.shape[0:2], np.uint8) * 255

plt.imshow(n_image)
```

### save image as png

```python
cv.imwrite('2.png', cv.cvtColor(n_image, cv.COLOR_RGBA2BGRA))
```
