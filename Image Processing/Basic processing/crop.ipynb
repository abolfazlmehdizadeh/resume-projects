### import packages

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### read image

```python
image = cv.imread('1.png')
```

### show image in RGB

```python
plt.imshow(image[:,:,::-1])
```

### get height and width

```python
H, W = image.shape[0:2]
```

### crop and show image

```python
crop_image = image[H//7 : H//3, W//2 : 3*W//5, ::-1]

plt.imshow(crop_image)
```
