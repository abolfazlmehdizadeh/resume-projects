```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
image = cv.imread('adad.jpg', 0)
```

```python
blur = cv.GaussianBlur(image, (15, 15), 0)
image = cv.adaptiveThreshold(blur, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
plt.imshow(image, cmap='gray')
```

```python
image = cv.bitwise_not(image)
plt.imshow(image, cmap='gray')
```

```python
num, label = cv.connectedComponents(image)
plt.imshow(label)
num
```
