```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
original_image = cv.imread('adad.jpg')
image = cv.imread('adad.jpg', 0)
blur = cv.GaussianBlur(image, (15, 15), 0)
image_b = cv.adaptiveThreshold(blur, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY_INV, 11, 2)
plt.imshow(image_b, cmap='gray')
```

```python
contours, _  = cv.findContours(image_b, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
```

```python
for i, j in enumerate(contours):
    cv.drawContours(original_image, [j], -1, (200, 20, 100), 2)
    M = cv.moments(j)
    x = M['m10'] / M['m00']
    y = M['m01'] / M['m00']
    cv.circle(original_image, (round(x), round(y)), 2, (255, 220, 0), 2)


rect = cv.boundingRect(contours[2])
cv.rectangle(original_image, rect, (0,0,255) ,1)
plt.imshow(original_image)
```
