```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
image = cv.imread('adad.jpg', 0)
```

```python
blur = cv.GaussianBlur(image, (13, 13), 0)
new_image = cv.adaptiveThreshold(blur, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
plt.imshow(new_image, cmap='gray')
```
